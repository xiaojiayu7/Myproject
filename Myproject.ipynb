```python
# 返回薪资上、下限值，默认返回最低薪资值
def get_Salary(str, method='bottom'):
    pos = str.find('-')                               # 包含“-”符号，则返回其索引位置，否则返回-1
    if pos != -1:
        botSalary = str[:pos-1]                       # 薪资下限值
        topSalary = str[pos+1:len(str)-1]             # 薪资上限值
    else:                                             # 针对“k以上”形式的值
        botSalary = str[:str.upper().find('K')]
        topSalary = botSalary
        
    if method == 'bottom':                            # 传入除了“bottom”以外的任意字符串，则返回topSalary
        return botSalary
    else:
        return topSalary
```

```python
df_duplicated['bottomSalary'] = df_duplicated.salary.apply(get_Salary).astype('int')              # 截取薪资下限值，并转换为整型数值
df_duplicated['topSalary'] = df_duplicated.salary.apply(get_Salary,method='top').astype('int')    # 截取薪资上限值，并转换为整型数值
df_duplicated
```

```python
df_duplicated['avgSalary'] = df_duplicated.apply(lambda x: (x.bottomSalary + x.topSalary) / 2,axis=1)
df_duplicated
```

```python
df_endData = df_duplicated[['city','companyShortName','companySize','education','positionName','positionLables','workYear','avgSalary']]
```

```python
bins = [0,3,5,10,15,20,30,100]
level = ['0-3','3-5','5-10','10-15','15-20','20-30','30+']
df_endData['level'] = pd.cut(df_endData['avgSalary'],bins=bins,labels=level)
df_endData[['avgSalary','level']]
```

```python
df_level = df_endData.groupby(['city','level']).avgSalary.count().unstack()      # 各个城市不同辛子端的岗位数量分布情况
df_level_prop = df_level.apply(lambda x:x/x.sum(),axis=1)                       # 某个城市的某个薪资段的职位数量与该城市所有职位数量的比值
df_level_prop.plot.bar(stacked=True,figsize=(13,8))
```

```python
first_s = df_endData.positionLables.str[1:-1]                      # str方法进行字符串截取
second_s = first_s.str.replace(' ','')                             # replace方法删除空格
third_s = second_s.dropna().str.split(',').apply(pd.value_counts)  # 以逗号作为分隔符，统计某个关键词在各个职位出现的次数
fourth_s = third_s.unstack().dropna().reset_index()                # 行列转换后后删除NaN值，并重置索引
fourth_s
```

```python
from wordcloud import WordCloud

fourth_s.index = fourth_s.level_0.str.replace("'",'')       # 删除单引号

wordcloud = WordCloud(font_path='C:\Windows\Fonts\STZHONGS.TTF',width=900,height=400,background_color='white')   # 初始化
f, axes = plt.subplots(figsize=(15,10))                     # 绘图区域
wordcloud.fit_words(fourth_s.level_1)                       # 根据词频生成词云
axes = plt.imshow(wordcloud)                                # 展示词云
plt.axis('off')                                             # 不显示x轴、y轴刻度
plt.show()
```
